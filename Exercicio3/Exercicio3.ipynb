```python
import networkx as nx
import pandas as pd
from IPython.display import display, HTML

%matplotlib inline
```

```python
G = nx.read_gml("./../data/GraphMissingEdges.gml")
```

```python
categories_nodes = nx.get_node_attributes(G,'categories')
for key in categories_nodes:
    categories_nodes[key] = categories_nodes[key].split(',')
```

```python
categories_df = pd.read_csv('./../data/categories.csv', sep=',')
categories_df = categories_df.set_index('CategoryId')
```

## 1- Usando a rede do desafio "GraphMissingEdges.gml", faça o que é pedido abaixo.
### a) Compute a centralidade de grau e identifique os top 50 lugares de acordo com essa métrica. Mostre o nome e categorias dos lugares.

```python
deg_cen = nx.degree_centrality(G)
nodes = sorted(deg_cen.items(), key=lambda x:x[1], reverse=True)[0:50]
degree_centrality = []
for node in nodes:
    categories = categories_nodes[node[0]]
    categories_names = []
    for category in categories:
        categories_names.append(categories_df.loc[int(category)]['names'])
    degree_centrality.append({'name': G.nodes[node[0]]['name'], 'categories': categories_names, 'degree_centrality': node[1]})

deg_cen_df = pd.DataFrame(degree_centrality)

display(HTML(deg_cen_df.to_html()))
```
